```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Income = [1350000,2150000,1570000,1730000,2010000,3210000,5540000]
Expenditure = [950000,1850000,1350000,1400000,2900000,3650000,1730000]
Year = [2005,2006,2007,2008,2009,2010,2011]

Surplus = []
for x in range(len(Income)):
    a = Income[x] - Expenditure[x]
    Surplus.append(a)

Surplus    
```

```python
plt.rcParams["figure.figsize"]=(7,7)
plt.rcParams["font.size"]=16
```

```python
plt.figure(figsize=(10,7))
# alpha = 0.3 makes the color faint
plt.plot(Year, Income, color = "black", label = "Income")
plt.plot(Year, Expenditure, color = "r", label = "Expenditure")
plt.plot(Year, Surplus, color = "b", marker = "x" ,label = "Surplus")
#plt.plot("k--*") color, linestyle and marker

plt.suptitle("Company XYZ Yearly Cashflow", fontsize = 27, fontweight = "bold", fontfamily = "algerian", color = "blue" )
plt.legend()
plt.minorticks_on()
plt.grid(which="both")
# which = both ,makes the grid smaller
```

```python
plt.figure(figsize=(10,7))

plt.bar(Year, Income, color = "g", label = "Income" )
plt.bar(Year, Expenditure, color = "r", label = "Expenditure", alpha = 0.3)
#plt.plot(Year, Surplus, color = "b", marker = "x" ,label = "Surplus")

plt.suptitle("Company XYZ Yearly Cashflow", fontsize = 27, fontweight = "bold", fontfamily = "algerian", color = "blue" )
plt.legend()
plt.minorticks_on()
plt.grid(which="both")
```

```python
plt.figure(figsize=(10,7))
# alpha = 0.3 makes the color faint
#plt.style.use("fivethrtyeight")
newx = np.arange(len(Year))
val = 0.3

plt.bar(newx-val, Income, width = val ,color = "g", label = "Income")
plt.bar(newx, Expenditure, width = val ,color = "r", label = "Expenditure")
plt.bar(newx+val, Surplus,width = val, color = "k" ,label = "Surplus")
#plt.plot("k--*") color, linestyle and marker

plt.xticks(ticks=newx, labels=Year)
plt.suptitle("Company XYZ Yearly Cashflow", fontsize = 27, fontweight = "bold", fontfamily = "algerian", color = "blue" )
plt.legend()
plt.minorticks_on()
plt.grid(which="both")
```

```python
plt.figure(figsize=(10,7))
# alpha = 0.3 makes the color faint
#plt.style.use("fivethrtyeight")
newx = np.arange(len(Year))
val = 0.3

plt.barh(newx-val, Income, height = val ,color = "g", label = "Income")
plt.barh(newx, Expenditure, height = val ,color = "r", label = "Expenditure")
plt.barh(newx+val, Surplus,height = val, color = "k" ,label = "Surplus")
#plt.plot("k--*") color, linestyle and marker

plt.yticks(ticks=newx, labels=Year)
plt.suptitle("Company XYZ Yearly Cashflow", fontsize = 27, fontweight = "bold", fontfamily = "algerian", color = "blue" )
plt.legend()
plt.minorticks_on()
plt.grid(which="both")
```
